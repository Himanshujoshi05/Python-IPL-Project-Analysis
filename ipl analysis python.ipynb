which team wins highest toss ?

```python
toss_wins = data['toss_winner'].value_counts()
color = ["orange","darkblue","blue","purple","lightblue","red","darkblue","yellow","darkred","pink"]

plt.figure(figsize=(10, 5)) # Set the figure size (10 means width and 5 means height)
bars = plt.bar(toss_wins.index,toss_wins.values,color=color,width=0.6)
plt.xlabel("Teams",fontsize=12)
plt.ylabel("No of toss wins.",fontsize=12)
plt.title("Most toss wins",fontsize=15)
plt.bar_label(bars, fmt='%d', fontsize=12) # Add labels to the bars , fmt ='%d' means display the value as a number

plt.show()
```

its is usefull for change bar label type like fmt  = '%f' for float values

%f: Floating-point number with six decimal places (default).
Example: 5.123456

%.1f: Floating-point number with one decimal place.
Example: 5.1

%s: String format (useful if labels contain text).
Example: 'Team A'

%.2f%%: Percentage with two decimal places (note the double %).
Example: 50.25%

percentage of choose field or bat

```python
# radius =1 to make it smaller,startangle = 90 to make it start from 90 degrees,autopct to display percentage on the pi,textprops to change the color of the text and manymore.
# family = 'sans-serif' to change the font family of the pie chart, more family option(serif,sans-serif,monospace,cursive,fantasy)
# counterclock = True to make the pie chart start from the right side of the circle,wedgeprops to change the color of the wedge of the pie chart 
# antialiased = True to make the pie chart look smoother,fill = True to fill the pie chart with color

toss_counts = data["toss_decision"].value_counts()

plt.pie(toss_counts,labels=toss_counts.index,autopct='%1.0f%%',startangle=90,shadow=True,colors=["pink","purple"],radius=1,
textprops={"fontsize":"x-large","color":"red","fontweight":"light","family":"cursive","rotation":0},counterclock=True,
wedgeprops={"edgecolor":"black","linewidth":2,"linestyle" : "--","antialiased":False,"fill":True},rotatelabels=2)
# plt.pie([1],colors=["white"],radius=0.7) this is optional to make the center of the pie chart white
plt.show()
```

total match played by teams

```python
total_matched_played = data["team1"].value_counts().add(data["team2"].value_counts(),fill_value=0).sort_values(ascending=False).astype(int) # In Pandas, .add() provides more flexibility than +, especially when handling missing values (NaN) or alignment issues.
total_matched_played
```

total match wins by teams

```python
total_match_wins = data["match_winner"].value_counts()
total_match_wins
```

winning % of all teams 


```python
winning_percentage = ((total_match_wins / total_matched_played) * 100).sort_values(ascending=False).astype(int)
winning_percentage
```

Create a DataFrame to store team performance data & Sort it by Win% (descending)

```python
team_performance = pd.DataFrame({
    "total matches played" : total_matched_played,
    "total match wins" : total_match_wins,
    "winning percentage (%)" : winning_percentage
}).sort_values(by='winning percentage (%)', ascending=False)
team_performance
```

```python
venue_avg_score = data.groupby("venue")[["first_ings_score","second_ings_score"]].mean().sort_values(by=["first_ings_score", "second_ings_score"],ascending=False).astype(int)
venue_avg_score
```

```python
venue_avg_score = venue_avg_score.sort_values(by='first_ings_score', ascending=True)

plt.figure(figsize=(8,5))
bars = plt.barh(venue_avg_score.index, venue_avg_score['first_ings_score'], color="pink")

plt.xlabel("Average score", fontsize=12)
plt.ylabel("Venue", fontsize=12)
plt.title("first inning Average score by venue", fontsize=15, fontweight="bold")
plt.bar_label(bars, fmt='%d', fontsize=12)

plt.show()
```

```python
venue_avg_score = venue_avg_score.sort_values(by='second_ings_score', ascending=True)

plt.figure(figsize=(8,5))
bars = plt.barh(venue_avg_score.index, venue_avg_score['second_ings_score'], color="pink")

plt.xlabel("Average score", fontsize=12)
plt.ylabel("Venue", fontsize=12)
plt.title("second inning Average score by venue", fontsize=15, fontweight="bold")
plt.bar_label(bars, fmt='%d', fontsize=12)

plt.show()
```

```python
higest_score = data.groupby(["team1", "team2"])[["first_ings_score", "second_ings_score"]].max().sort_values(by=["first_ings_score", "second_ings_score"], ascending=False).head(10)
higest_score # The double square brackets are used to select the columns first_ings_score and second_ings_score from the grouped data. This creates a new DataFrame containing only these two columns.
```

```python
# Reset the index to make it a regular DataFrame
higest_score_reset = higest_score.reset_index()

# Create a figure with a specified size
plt.figure(figsize=(15, 10))

# Plot the first innings scores, shifted slightly to the left
bars = plt.bar(higest_score_reset.index - 0.2, higest_score_reset['first_ings_score'], width=0.4, label='First Innings Score')

# Plot the second innings scores, shifted slightly to the right
bars1 = plt.bar(higest_score_reset.index + 0.2, higest_score_reset['second_ings_score'], width=0.4, label='Second Innings Score')

# Add labels and title
plt.xlabel('Match', fontsize=12, fontweight="bold")
plt.ylabel('Score', fontsize=12, fontweight="bold")
plt.title('Highest Scores in First and Second Innings', fontsize=16, fontweight="bold")
plt.xticks(higest_score_reset.index, higest_score_reset['team1'] + ' vs ' + higest_score_reset['team2'], rotation=45, ha='right', fontsize=12, fontweight="normal")
plt.bar_label(bars, fmt='%d', fontsize=12, color="red")
plt.bar_label(bars1, fmt='%d', fontsize=12)

# Add legend
plt.legend(loc="upper right")

# Display the plot
plt.tight_layout()  # Adjust the plot to ensure everything fits without overlapping
plt.show()
```

```python
player_of_the_match = data["player_of_the_match"].value_counts().head(10)
player_of_the_match
```

```python
player_of_the_match = player_of_the_match.sort_values(ascending=True)
plt.figure(figsize=(8, 5))

bars = plt.barh(player_of_the_match.index,player_of_the_match.values,color="pink")

plt.xlabel("No of times",fontsize=12)
plt.ylabel("Players",fontsize=12)
plt.title("Top 10 players of the match",fontsize=15,fontweight="bold")
plt.bar_label(bars,fmt='%d',fontsize=12)
plt.xticks(np.arange(0, 6, 1))  

plt.show()
```

```python

# Group by player and match_winner, count the occurrences, and sort to get the top 10 players
top_players = data.groupby(['player_of_the_match', 'match_winner']).size().reset_index(name='count').sort_values(by='count', ascending=False).head(10)
# Counts the number of occurrences for each group (i.e., how many times each player won the Player of the Match award).
# Converts the grouped data into a DataFrame with a new column named 'count'.
top_players = top_players.sort_values(ascending=True,by='count')

# Create a figure with a larger size
plt.figure(figsize=(12, 8))

# Create a bar chart
bars = plt.barh(top_players['player_of_the_match'], top_players['count'], color='skyblue')

# Add labels and title
plt.xlabel('Number of Awards')
plt.ylabel('Player')
plt.title('Top 10 Players of the Match')
plt.xticks(np.arange(0,6,1))

# Add team names as annotations # yah code nhi aaya samaj mai
for bar, team in zip(bars, top_players['match_winner']):
    x_pos = bar.get_width()  # Get the number of awards (bar length)
    y_pos = bar.get_y() + bar.get_height() / 2  # Find the center of the bar
    plt.text(x_pos, y_pos, team, va='center', ha='left')  # Add the team name


# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```
